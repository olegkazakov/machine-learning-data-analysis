```python
import numpy as np
import math
```

$$ f(x) = \sin ({x}/{5}) * \exp ({x}/{10}) + 5 * \exp ({-x}/{2}) $$

```python
# base function

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
```

```python
# search of solution in dots (1, 15)
a = np.array([[1., 1.], [15., 1.]])
b = np.array([f(1.), f(15.)])
w = np.linalg.solve(a, b)
print a, b, w

def f1(x):
    return w[0]*x + w[1] # уравнение прямой
```

```python
# search of solution in dots (1, 8, 15)
a = np.array([[1., 1., 1.], [8**2., 8., 1.], [15.**2, 15., 1.]])
b = np.array([f(1.), f(8.), f(15.)])
s = np.linalg.solve(a, b)
print a, b, s

def f2(x):
    return s[0]*x**2 + s[1]*x + s[2] # уравнение параболы
```

```python
# search of solution in dots (1, 4, 10, 15)
a = np.array([[1., 1., 1., 1.], [4.**3, 4.**2, 4., 1.], [10.**3, 10.**2, 10., 1.], [15.**3, 15.**2, 15., 1.]])
b = np.array([f(1.), f(4.), f(10.), f(15.)])
r = np.linalg.solve(a, b)
print a, b, r

def f3(x):
    return r[0]*x**3 + r[1]*x**2 + r[2]*x + r[3] # уравнение кубической параболы
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
% matplotlib inline
mpl.rc('font', family='Verdana', size =16)
fig, ax = plt.subplots(figsize=(10,5))
x = np.linspace(1, 15, 200)

ax.axis([0., 16., 0., 3.5])
ax.grid()
ax.plot(x, f(x), label = u'Исходная функция')
ax.plot(x, f1(x), label = u'Многочлен 1-й степени')
ax.plot(x, f2(x), label = u'Многочлен 2-й степени')
ax.plot(x, f3(x), label = u'Многочлен 3-й степени')
ax.set_xlabel(u'x',{'fontname':'Arial', 'size': 24})
ax.set_ylabel(u'f(x)',{'fontname':'Arial', 'size': 24})
plt.plot([1, 4, 8, 10, 15], [f(1.), f(4.), f(8.), f(10.), f(15.)], 'ro', label = u'Точки для\nпостроения графика')

ax.legend(bbox_to_anchor=(1.7, 1.))

plt.show()
```
